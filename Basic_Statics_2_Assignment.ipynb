```python
from scipy import stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
printer_data = pd.Series([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
```

```python
printer_data
```

```python
printer_ci = stats.norm.interval(0.99, loc=printer_data.mean(), scale=printer_data.std())
print('Mean number of characters to be printed in range of', np.round(printer_ci, 4))
```

```python
plt.axvline(printer_data.mean(), color='red', linestyle='--', label='Sample Mean')
plt.axvline(printer_ci[0], color='blue', linestyle='dotted', label='Lower CI Bound')
plt.axvline(printer_ci[1], color='blue', linestyle='dotted', label='Upper CI Bound')
plt.hlines(y=1, xmin=printer_ci[0], xmax=printer_ci[1], color='skyblue', linewidth=4, label='99% Confidence Interval')
plt.xlabel('Value')
plt.title('Confidence Interval Visualization using stats.norm.interval')
plt.legend()
plt.show()
```
